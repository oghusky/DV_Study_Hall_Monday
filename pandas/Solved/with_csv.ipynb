# Filter Data

```python
poison_grass = df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Poison")]
poison_grass
#prints where type 1 is grass and type 2 is poison
```

```python
poison_or_grass = df.loc[(df["Type 1"]=="Grass") | (df["Type 2"]=="Poison")]
poison_or_grass
#prints if type 1 is grass or type 2 equals poison
```

```python
high_poison_grass = df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Poison") & (df["HP"] > 70)]
high_poison_grass
```

```python
high_poison_grass.reset_index()
#gives new index values to filtered data
```

```python
high_poison_grass.reset_index(drop=True)
#drop equals true removes index column
```

```python
#filter all names that contain the word "Mega"
df.loc[df["Name"].str.contains("Mega")]
```

```python
#filter all names that don't contain the word "Mega"
df.loc[~df["Name"].str.contains("Mega")]
```

```python
import re
#filter using regEx
#flags=re.I << ignore casing of letters
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I,regex=True)]
```

```python
df.loc[df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)]
```

```python
#locate where type 1 equals fire and change it to
df.loc[df["Type 1"] == "Fire", "Type 1"] = "Dragon"
df
```
