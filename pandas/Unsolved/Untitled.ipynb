# Working with APIs

```python
import requests
import pprint
pp = pprint.PrettyPrinter(width=41, compact=True)
# import pandas as pd
```

```python
url = "https://api.chucknorris.io/jokes/random"
# response = requests.get(url)
response = requests.get(url).json()
pp.pprint(response)
```

```python
list_of_jokes = []
def more_jokes(var):
    return requests.get(var).json()
for joke in range(3):
    list_of_jokes.append(more_jokes(url))
pp.pprint(list_of_jokes)
```
